### Filter the dataframe with only this columns ['continent','location','date','total_cases','total_deaths','gdp_per_ca

```python
df.head(1)
```

```python
df1 = df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
df1.head()
```

#### Data Cleaning
#### Remove all duplicates observations
#### b. Find missing values in all columns
#### c. Remove all observations where continent column value is missing
#### Tip : using subset parameter in dropna
#### d. Fill all missing values with 0

```python
df1.duplicated().sum()
```

### There are not duplicates in this dataset

```python
df1.drop_duplicates()
```

```python
df1.isnull().sum()
```

#### Removing all missing values from continent column

```python
# df1['continent'].dropna()
df1.dropna(subset=["continent"])
```

#### Filling all missing values with 0

```python
df1 = df1.fillna(0)
```

```python
df1.isnull().sum()
```

```python
df.isnull().sum()
```

#### Date time format :
#### a. Convert date column in datetime format using pandas.to_datetime

```python
df1.head(1)
```

```python
df1['date'] = pd.to_datetime(df1['date'])
```

```python
df1.dtypes
```

#### b. Create new column month after extracting month data from date column.

```python
df1['Month'] = df1['date'].dt.month
```

```python
df1.head(1)
```

#### Data Aggregation:
#### a. Find max value in all columns using groupby function on 'continent' column
#### Tip: use reset_index() after applying groupby

```python
df1.groupby('continent').max().reset_index()
```

#### b. Store the result in a new dataframe named 'df_groupby'.
#### (Use df_groupby dataframe for all further analysis)

```python
df_groupby = df1.groupby('continent').max().reset_index()
```

```python
df_groupby.head()
```

#### Feature Engineering :
#### a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'

```python
df_groupby['total_deaths_to_total_cases'] = (df_groupby['total_deaths']/df_groupby['total_cases'])*100
```

```python
df_groupby.head()
```

```python
df_groupby['total_deaths_to_total_cases'].head()
```
